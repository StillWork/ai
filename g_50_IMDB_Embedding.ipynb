# 단어 임베딩
### Embedding 계층을 사용하여 쉽게 만들 수 있다
- 정수 인덱스를 벡터로 매핑하는 딕셔너리 구조 (인덱스 크기, 벡터 크기)
- 학습 시키는 데이터에 따라 다른 임베딩이 만들어진다.

### IMDB 영화 리뷰 데이터를 사용한 임베딩 예제



```python
from keras.models import Sequential
from keras.layers import Flatten, Dense, Embedding
import os, os.path
import zipfile
from keras.datasets import imdb
from keras import preprocessing
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
import numpy as np
import matplotlib.pyplot as plt
```

```python
max_features = 10000
maxlen = 20
(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=max_features)
```

```python
x_train.shape
```

```python
len(x_train[0])
```

```python
# 각 문장이 몇개의 단어로 구성되어 있는지 확인
for i in range(10):
    print(len(x_train[i]))
```

```python
x_train[:3]
```

```python
# 뒤에서부터 20개의 단어만 사용한다.
x_train_p=preprocessing.sequence.pad_sequences(x_train, maxlen=maxlen)
x_test_p=preprocessing.sequence.pad_sequences(x_test, maxlen=maxlen)
```

```python
x_train_p[:3]
```

```python
model = Sequential()
model.add(Embedding(10000, 8, input_length=maxlen))
model.add(Flatten())
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy',
              optimizer='rmsprop',
              metrics=['acc'])
model.summary()
```

```python
history = model.fit(
    x_train_p, y_train,
    epochs=10,
    batch_size=32,
    validation_split=0.2)
```

## 위의 결과는 20개의 단어만 고려한 것임
### 성능이 75% 정도 됨
- 각 단어를 독립적으로 다루었으며, 문장의 구성 정보를 고려하지 않음
- 문장의 구조 정보를 고려하려면 임베딩 층 위에 합성곱이나 순환신경망 층을 추가한다
